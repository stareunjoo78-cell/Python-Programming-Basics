```python
# folium은 Leaflet.js를 이용해서 위치 좌표계 값을 지도위 시각화 할 수 있다.
import folium
```

```python
# folium의 Map 클래스의 location 속성으로 지정한 위치를 중심으로 하는 zoom_start 속성으로 지정한 배율을
# 사용하는 지도를 만든다.
# folium.Map(location=[위도, 경도], zoom_start=배율)
folium_map = folium.Map(location=[37.5697599, 126.9836604], zoom_start=16)
folium_map
```

```python
folium_map = folium.Map(location=[37.5697599, 126.9836604], zoom_start=15)

# folium의 Marker 클래스의 location 속성으로 지정한 위치에 마커를 만든다.
# 마커는 add_to() 메소드로 지도에 추가한다.
# folium의 Marker 클래스의 popup 속성으로 마커를 클릭했을 때 표시할 문자열을 지정할 수 있다.
folium.Marker(location=[37.57961628, 126.97704096], popup='경복궁').add_to(folium_map)
folium.Marker(location=[37.56580575, 126.97514357], popup='덕수궁').add_to(folium_map)
folium.Marker(location=[37.5787512, 126.9948775], popup='창경궁').add_to(folium_map)
folium.Marker(location=[37.57943211, 126.99104354], popup='창덕궁').add_to(folium_map)
folium.Marker(location=[37.57373256, 126.9937536], popup='종묘').add_to(folium_map)

# 마커를 클릭했을 때 팝업되는 말풍선 크기는 folium의 Popup 클래스의 max_width 속성으로 지정할 수 있다.
popup = folium.Popup('영풍문고 - 종로본점', max_width=200)
folium.Marker(location=[37.56978693, 126.98231618], popup=popup).add_to(folium_map)

# save() 메소드로 지도를 html 파일로 저장할 수 있다.
folium_map.save('./output/folium_map.html')
folium_map
```

```python
folium_map = folium.Map(location=[37.5697599, 126.9836604], zoom_start=15)

# folium의 Icon 클래스의 color 속성으로 마커의 색상과 icon 속성으로 아이콘 모양을 지정할 수 있다.
# 마커 색생과 아이콘 모양을 변경하려면 folium의 Marker 클래스의 icon 속성에 Icon 객체를 지정하면 된다.
# 색상 이름 리스트
# 'darkred', 'pink', 'gray', 'cadetblue', 'purple', 'lightgray', 'red', 'beige', 'orange', 'blue', 'black', 
# 'white', 'darkblue', 'lightgreen', 'lightblue', 'darkpurple', 'darkgreen', 'green', 'lightred'
# 마커에 표시할 아이콘은 https://getbootstrap.com/docs/3.3/components/#glyphicons-glyph에서 제공하는 
# 아이콘을 표시할 수 있다.

folium.Marker(location=[37.57961628, 126.97704096], popup='경복궁',
             icon=folium.Icon(color='pink', icon='ok')).add_to(folium_map)
folium.Marker(location=[37.56580575, 126.97514357], popup='덕수궁',
             icon=folium.Icon(color='beige', icon='remove')).add_to(folium_map)
folium.Marker(location=[37.5787512, 126.9948775], popup='창경궁',
             icon=folium.Icon(color='orange', icon='tag')).add_to(folium_map)
folium.Marker(location=[37.57943211, 126.99104354], popup='창덕궁',
             icon=folium.Icon(color='cadetblue', icon='arrow-up')).add_to(folium_map)
folium.Marker(location=[37.57373256, 126.9937536], popup='종묘',
             icon=folium.Icon(color='purple', icon='hand-down')).add_to(folium_map)

popup = folium.Popup('영풍문고 - 종로본점', max_width=200)
folium.Marker(location=[37.56978693, 126.98231618], popup=popup,
             icon=folium.Icon(color='green', icon='chevron-down')).add_to(folium_map)

folium_map.save('./output/folium_map2.html')
folium_map
```

```python
folium_map = folium.Map(location=[37.5697599, 126.9836604], zoom_start=15)

# folium의 CircleMarker 클래스로 원형 마커를 만들 수 있다.
# radius 속성으로 마커로 표시되는 원의 반지름을 지정한다.
# color 속성으로 원의 선 색상을 fill_color 속성으로 원 내부의 채울 색상을 지정할 수 있다.

folium.CircleMarker(location=[37.57961628, 126.97704096], popup='경복궁',
                   radius=100, color='red', fill_color='blue').add_to(folium_map)
popup = folium.Popup('영풍문고 - 종로본점', max_width=200)
folium.CircleMarker(location=[37.56978693, 126.98231618], popup=popup, 
                    radius=50, color='#00ffff', fill_color='#ff00ff').add_to(folium_map)

folium_map.save('./output/folium_map3.html')
folium_map
```
